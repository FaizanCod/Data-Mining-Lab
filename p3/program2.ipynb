#### Cosine Similarity between documents

```python
import pandas as pd
import numpy as np
```

```python
doc_cnt = 5
```

```python
for i in range(doc_cnt):
    filename = f"file{i}.txt"
    with open(filename, 'r') as file:
        continue
        
```

```python
import string
stopWordsFile = open('stopWords.txt', 'r')
stopWords = []
for word in stopWordsFile:
    stopWords.append(word.rstrip('\n\r'))
```

```python
distinct = []
freq = []
remChar = string.punctuation.replace("'", '')
for i in range(doc_cnt):
    filename = f"file{i}.txt"
    content = open(filename, 'r').read().lower()
    content = content.translate(str.maketrans('', '', remChar)).split()
    currFreq = {}
    for word in content:
        if(word not in stopWords):
            if(word not in distinct): distinct.append(word) #storing distinct words in all the docs
            if(currFreq.__contains__(word)): currFreq[word]+=1 #storing freq of each word in the current doc
            else: currFreq[word] = 1
    freq.append(currFreq) #making an array of dict of words with freq in each doc

distinct
```

```python
docName = []
for docIdx in range(5):
    docName.append(f'file_{docIdx}')
    for word in distinct:
        if(not freq[docIdx].__contains__(word)): freq[docIdx][word] = 0 #placing words with 0 freq in each doc dict

data = {'Name': docName}
Cosine_df = pd.DataFrame(data)
Cosine_df.set_index('Name', inplace=True)
```

```python
import numpy as np
for docId1 in range(doc_cnt):
    cosine_val = []
    for docId2 in range(doc_cnt):
        if(docId1 >= docId2): cosine_val.append(np.nan)
        else:
            dot, mod_x, mod_y = 0, 0, 0
            for word in distinct:
                dot+=freq[docId1][word]*freq[docId2][word]
                mod_x+=freq[docId1][word]**2
                mod_y+=freq[docId2][word]**2
            mod_x = mod_x**(1.0/2)
            mod_y = mod_y**(1.0/2)
            cosine_val.append(dot/(mod_x*mod_y))
    # print(cosine_val)
    Cosine_df[f'file_{docId1}'] = cosine_val

Cosine_df
```
