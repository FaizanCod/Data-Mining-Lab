```python
import pandas as pd
import math
```

```python
data = pd.read_csv('data.csv')
items = data['Items']
size = len(items)

support = 0.5
min_sup = support * size
data

dataset = {}
for i, row in data.iterrows():
    transaction_id = row['Transaction']
    items = row['Items'].split(',')
    dataset[transaction_id] = items

# print(dataset)
```

```python
def calc (items, min_sup):
    c1 = {}
    for itemset in items:
        for item in itemset:
        # print(it[j])
            if item not in c1:
                c1[item] = 1
            else:
                c1[item] += 1
    print(c1)

    l1 = {item: c for item, c in c1.items() if c >= min_sup}
    return l1
```

```python
def initialize (data):
    itemset = [{item: 0 for item in items} for items in data]
    print(itemset)
    return itemset

def apriori(data, min_sup):
    frequent = {}
    itemsets = initialize(data)

    i=1
    while itemsets:
        freq_items = calc(itemsets, min_sup)
        # if no more items can be made
        if not freq_items:
            break
        frequent[i] = freq_items

        i+=1
        itemsets = [{it for it in data_item if it in freq_items} for data_item in data]
    
    return frequent
```

```python
freq_items = apriori(dataset, min_sup)
print("Frequent items")
for i, freq in freq_items.items():
    print(f"Size {i}: ", freq)
```
