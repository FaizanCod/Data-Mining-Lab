```python
import numpy as np
import pandas as pd

my_data = pd.read_csv('animal.txt')
animal_names = my_data['Unnamed: 0']
my_data
```

```python
# handling NaN values in cells by taking mean of column and rounding off
my_data['end'].fillna(round(my_data['end'].mean()), inplace=True)
my_data['gro'].fillna(round(my_data['gro'].mean()), inplace=True)
```

```python
my_data
```

```python
# creating a dist matrix of data points * data points
dist_matrix = np.zeros((len(my_data), len(my_data)))
dist_matrix.shape
```

```python
import math
# for each of the data points, we take the lower triangular matrix to store the euclidean distance of attributes for ith and jth data points
for i in range(0, len(dist_matrix)):
    for j in range(0, i):
        temp = 0
        # for each of the attributes in a data point
        for k in range(1, len(my_data.iloc[0])):
            # squaring
            temp += (my_data.loc[i][k] - my_data.loc[j][k]) ** 2
        # storing the sqrt of the square
        dist_matrix[i][j]= math.sqrt(temp)
```

```python
# to display as a dataframe
distance = pd.DataFrame(dist_matrix, animal_names, animal_names)
distance
```

```python
# dictionary to store the pairwise difference
data_dict = {}
for i in range(0,len(distance)):
    for j in range(0,i):
        data_dict[animal_names[i] + "-" + animal_names[j]] = dist_matrix[i][j]

# data_dict
```

```python
# Sort data_dict based on the values (ascending)
sorted_data_dict = sorted(data_dict.items(), key=lambda x: x[1])
# showing only 10 most similar animals (lowest difference)
sorted_data_dict[:10]
```
